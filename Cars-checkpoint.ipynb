```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "CSV_Files/cars.csv"
```

```python
df = pd.read_csv(file, low_memory=False)
df.head()
```

```python
df.count()
```

```python
df.dtypes
```

```python
df.rename(columns={'Dimensions.Height': 'Height',
                   'Dimensions.Length': 'Length',
                   'Dimensions.Width':'Width',
                   'Engine Information.Driveline': 'Drivetrain',
                   'Engine Information.Engine Type':'Engine_Type',
                   'Engine Information.Hybrid' : 'Hybrid',
                   'Engine Information.Number of Forward Gears': 'Gears',
                   'Engine Information.Transmission': 'Transmission',
                   'Fuel Information.City mpg' : 'City_MPG',
                   'Fuel Information.Highway mpg' : 'Highway_MPG',
                   'Fuel Information.Fuel Type': 'Fuel_Type',
                   'Identification.Classification': 'Cassification', 
                   'Identification.ID': 'Car Name',
                   'Identification.Make': 'Make',
                   'Identification.Model Year' : 'Model',
                   'Identification.Year' : 'Year',
                   'Engine Information.Engine Statistics.Horsepower': 'Horsepower',
                   'Engine Information.Engine Statistics.Torque': 'Torque'},  inplace=True)
df
```

```python
#Average MPG by Make
new_df = df.set_index("Make")
new_df.head()
```

```python
new_df[['City_MPG', 'Highway_MPG']]
```

```python
city_average = new_df['City_MPG'].mean()
highway_average = new_df['Highway_MPG'].mean()
city_average,highway_average
```

```python
mpgs = new_df.groupby('Make')
# mpgs

avg_city = mpgs.mean()["City_MPG"]
#avg_city

avg_highway = mpgs.mean()["Highway_MPG"]
#avg_highway

total_avg = (avg_city + avg_highway) / 2

avg_mpg = pd.DataFrame({"Average City MPG": avg_city, 
                        "Average Highway MPG": avg_highway,
                       "Total Average MPG" : total_avg})
#avg_mpg

avg_mpg.index.name = 'Make'
avg_mpg
```

```python
fig = plt.figure(figsize=(20,15))
plt.xticks(rotation=90)
plt.plot(avg_mpg)
```

```python
horsepower_df = df.groupby(['Drivetrain'])
#horsepower_df.count().head(10)

horsepower = horsepower_df["Horsepower"].mean()
#horsepower

torque = horsepower_df["Torque"].mean()
#torque

hoursepower_torque = pd.DataFrame({"Average Horsepower": horsepower, 
                           "Average Torque": torque})

hoursepower_torque.index.name = 'Drivetrain'
hoursepower_torque
```

```python
plt.plot(hoursepower_torque)
```

```python
drivetrain_df = df.groupby(['Drivetrain'])
#drivetrain_df.count().head(10)

city_mpg = drivetrain_df["City_MPG"].mean()
city_mpg

highway_mpg = drivetrain_df["Highway_MPG"].mean()
#highway_mpg

drivetrain = pd.DataFrame({"Average City MPG": city_mpg, 
                           "Average Highway MPG": highway_mpg})
#avg_mpg

drivetrain.index.name = 'Drivetrain'
drivetrain
```

```python
plt.plot(drivetrain)
```

```python
year = df.groupby(['Year'])
year.count().head(10)

city_mpg = year["City_MPG"].mean()
#city_mpg

highway_mpg = year["Highway_MPG"].mean()
#highway_mpg

year_mpg = pd.DataFrame({"Average City MPG": city_mpg, 
                         "Average Highway MPG": highway_mpg})
year_mpg

year_mpg.index.name = 'Year'
year_mpg
```

```python
plt.legend()
plt.plot(year_mpg)
```
