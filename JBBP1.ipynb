```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "CSV_Files/cars.csv"
```

```python
makeandsize_df = pd.read_csv(file, low_memory=False)
makeandsize_df.head()
```

```python
makeandsize_df.rename(columns={'Identification.Make': 'Make',
                    'Dimensions.Height': 'Height',
                   'Dimensions.Length': 'Length',
                   'Dimensions.Width':'Width',
                   'Engine Information.Driveline': 'Drivetrain',
                   'Engine Information.Engine Type':'Engine_Type',
                   'Engine Information.Hybrid' : 'Hybrid',
                   'Engine Information.Number of Forward Gears': 'Gears',
                   'Engine Information.Transmission': 'Transmission',
                   'Fuel Information.City mpg' : 'City_MPG',
                   'Fuel Information.Highway mpg' : 'Highway_MPG',
                   'Fuel Information.Fuel Type': 'Fuel_Type',
                   'Identification.Classification': 'Classification', 
                   'Identification.ID': 'Car Name',
                   'Identification.Make': 'Make',
                   'Identification.Model Year' : 'Model',
                   'Identification.Year' : 'Year',
                   'Engine Information.Engine Statistics.Horsepower': 'Horsepower',
                   'Engine Information.Engine Statistics.Torque': 'Torque'},  inplace=True)
makeandsize_df
```

```python
#Googled to https://bit.ly/3yuOykQ
makeandsize_df['Size'] = makeandsize_df.Height * makeandsize_df.Length * makeandsize_df.Width
makeandsize_df['Average_MPG']=makeandsize_df.City_MPG+makeandsize_df.Highway_MPG/2
makeandsize_df = makeandsize_df.set_index("Size")
makeandsize_df.head()
```

```python
makeandsize_df[['City_MPG', 'Highway_MPG','Average_MPG']]
```

```python
size = makeandsize_df.groupby(['Size'])
size.count().head(10)

city_mpg = size["City_MPG"].mean()
#city_mpg

highway_mpg = size["Highway_MPG"].mean()
#highway_mpg

overall_mpg=size["Average_MPG"].mean()

size_mpg = pd.DataFrame({"Average City MPG": city_mpg, 
                         "Average Highway MPG": highway_mpg,
                        "Average_MPG":overall_mpg})
size_mpg

size_mpg.index.name = 'Size Inch^3'
size_mpg
```
